```python
# Step 1: Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np



# Load the dataset
df = pd.read_csv("/content/sample_data/Mental_Health_and_Social_Media_Balance_Dataset.csv")
df.head()
df.columns
```

```python
var=df["Gender"]
var.tail()
```

```python
df = df.rename(columns={"Daily_Screen_Time(hrs)": "Social Media Usage"})

```

## Step 2: Line Chart
Let's start by creating a simple line chart using `sns.lineplot()`. We'll plot one of the numeric columns in the dataset.

```python
# Simple line plotDaily_Screen_Time(hrs)
sns.lineplot(x='Age', y='Exercise_Frequency(week)', data=df)
plt.show()
```

## Step 3: Adding `hue` to Line Chart
Now, let's add a `hue` argument to separate the lines by a categorical variable (e.g., 'Gender').

```python
# Line plot with hue
sns.lineplot(x='Age', y='Happiness_Index(1-10)', hue='Gender', data=df, ci=None)
plt.title('Social Media Usage vs Age by Gender')
plt.show()
```

## Step 4: Barplot
Next, let's create a bar plot using `sns.barplot()`. This is useful for comparing categorical data.

```python
# Bar plot
sns.barplot(x='Gender', y='Exercise_Frequency(week)', data=df)
plt.title('Average Social Media Usage by Gender')
plt.show()
```

## Step 5: Boxplot
A boxplot helps in visualizing the distribution of data. Let's create one using `sns.boxplot()`.

```python
# Box plot
sns.boxplot(x='Gender',y="Exercise_Frequency(week)", data=df)
plt.title('Social Media Usage Distribution by Gender')
plt.show()
```

## Step 6: Heatmap
A heatmap is used to visualize correlations or data matrices. We'll create one using `sns.heatmap()`.

```python
# Ensure that only numeric columns are considered for the correlation matrix
numeric_data = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## Step 7: Violin Plot
Violin plots are a combination of boxplot and KDE. Let's create one using `sns.violinplot()`.

```python
# Violin plot
var= sns.load_dataset("tips")
sns.violinplot(x='smoker', y='tip', data=var, hue="smoker")

plt.title('Social Media Usage by Gender (Violin Plot)')
plt.show()
```

## Step 8: Scatter Plot
Scatter plots are used to observe relationships between two numeric variables. Let's create one using `sns.scatterplot()`.

```python
# Scatter plot
sns.scatterplot(x='total_bill', y='tip', hue="sex", data=var)
plt.title('Social Media Usage vs Age (Scatter Plot)')
plt.show()
```

## Step 9: Count Plot
Count plots are useful for displaying the frequency of categories. Let's create one using `sns.countplot()`.

```python
# Count plot
sns.countplot(x='Gender', data=df)
plt.title('Count of Gender')
plt.show()
```

## Step 10: Pie Chart
Although Seaborn doesn't support pie charts directly, we can use `matplotlib` to create one.

```python
# Pie chart
gender_counts = df['Gender'].value_counts()
gender_counts.plot.pie(autopct='%1.1f%%', figsize=(8,8))
plt.title('Gender Distribution (Pie Chart)')
plt.ylabel('')  # Hide the y-axis label
plt.show()
```
